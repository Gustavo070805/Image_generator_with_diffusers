```python
from diffusers import StableDiffusionPipeline
import torch

# Carregar o pipeline do Stable Diffusion
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipe.to("cuda")  # Use "cuda" se houver uma GPU disponível, senão use "cpu"

# Descrição da imagem a ser gerada
prompt = (
    "A cyber dragon flying"

)

# Parâmetros de ajuste avançados
num_inference_steps = 150  # Aumentar o número de passos para gerar mais detalhes
guidance_scale = 8.5  # Tornar a imagem ainda mais fiel ao prompt
height = 768  # Aumentar a resolução da imagem (altura)
width = 768  # Aumentar a resolução da imagem (largura)

# Gerar a imagem com parâmetros avançados
image = pipe(prompt, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale, height=height, width=width).images[0]

# Exibir a imagem gerada
image.show()

# Salvar a imagem em um arquivo PNG
image.save("ImageGenerator.png")

print("Imagem de alta qualidade gerada e salva com sucesso como ImageGenerator.png!")
```
